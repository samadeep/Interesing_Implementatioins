```python
#Perlin Noise Implementation and its graph

import random

class Perlin:
    
    def __init__( self ):
        self.gradients= []
        self.lowerBound = 0  

        
        
        #keeping 0 as lower_bound for the objects so if someone tries on a value of -negatives the out of bound error shows 
        
    def valueAt ( self , t ):
       
        if( t < self.lowerBound ):
            print(" Error !!! Out of Bounds ")
            return
        
 
        
        # Uniformly Assign gradients g1 g2 g3 g4 ..... gn to all the points of the coordinates
        
        while t >= len(self.gradients) - 1 + self.lowerBound:
            self.gradients.append(random.uniform(-1,1))
            
        discarded = int(self.lowerBound)
        
        
        d1 = ( t - t // 1)
        
        d2 = d1 - 1
        
        a1 = self.gradients[(int)(t//1)-discarded]*d1
        a2 = self.gradients[(int)(t//1+1)-discarded]*d2
        
        amt = self.__ease(d1)
        # smoothening curve used here --->
        
        
        return self.__lerp(a1,a2,amt)
        
        
    def discard(self, amount):
        gradientsToDiscard = int( amount+ self.lowerBound%1 )
        self.gradients = self.gradients[gradientsToDiscard:]
        self.lowerBound += amount
        
        
    def __ease(self,x):
        return 6*x**5-15*x**4+10*x**3
    
    def __lerp( self , start , stop , amt ):
        return amt*(stop-start)+start
        
      
        
        
```

```python
import matplotlib.pyplot as plt
```

```python
 noise = Perlin()
```

```python
time = [ i*0.01 for i in range(600) ]
values = [ noise.valueAt(i) for i in time]
values
```

```python
plt.title("Perlin Noise")
plt.xlabel("Time")
plt.ylabel("Value")

plt.plot(time, values)
plt.show()
```
